```python
# cv2, numpy, matplotlib.pyplot
import cv2 # 컴퓨터비전용 라이브러리
import numpy as np # 행렬 연산
import matplotlib.pyplot as plt # 시각화 라이브러리
```

```python
img = cv2.imread("images/Lenna.png", cv2.IMREAD_GRAYSCALE) # RGB -> BGR
h,w = img.shape
```

```python
# 3. 원(circle) 모양의 mask 생성하고 출력하기

center_h, center_w = h//2, w//2 # 정수값으로 반환
radius = 80
color = 255
thickness = -1
mask= np.zeros((h, w), np.uint8) # 검정색 사각형
cv2.circle(mask, (center_h, center_w), radius, color, thickness) # 안에 하얀 원

plt.imshow(mask, cmap = 'gray')
```

```python
# # 4. cv2를 사용하여 논리 연산 수행하기
# cv2.bitwise_and(img, mask) : AND 연산
# cv2.bitwise_or(img, mask) : OR 연산
# cv2.bitwise_xor(img, mask) : XOR 연산
# cv2.bitwise_not(img, mask) : NOT 연산

ex3_result1 = cv2.bitwise_and(img, mask)
ex3_result2 = cv2.bitwise_or(img, mask)
ex3_result3 = cv2.bitwise_xor(img, mask)
ex3_result4 = cv2.bitwise_not(img, mask)
```

```python
plt.subplot(2, 3, 1); plt.imshow(img, cmap = 'gray')
plt.subplot(2, 3, 2); plt.imshow(ex3_result1, cmap = 'gray')
plt.subplot(2, 3, 3); plt.imshow(ex3_result2, cmap = 'gray')
plt.subplot(2, 3, 4); plt.imshow(ex3_result3, cmap = 'gray')
plt.subplot(2, 3, 5); plt.imshow(ex3_result4, cmap = 'gray')
```
