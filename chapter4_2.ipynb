```python
# cv2, numpy, matplotlib.pyplot
import cv2 # 컴퓨터비전용 라이브러리
import numpy as np # 행렬 연산
import matplotlib.pyplot as plt # 시각화 라이브러리
import math
```

```python
img = cv2.imread("images/Lenna.png", cv2.IMREAD_GRAYSCALE) # RGB -> BGR
h,w = img.shape
print(h,w)
```

```python
# 이중 반복문을 활용하여, 영상(배열)의 각 픽셀(pixel)에 접근하고 ‘×1.7’을 수행하라 
# 랩핑
# ex2_result = img.copy() # 똑같은 영상을 복사
ex2_result = np.zeros((h,w)) # 초기화 되어 있다
for i in range(0, h):
    for j in range(0,w):
        if img[i,j] * 1.7 <= 255:
            ex2_result[i,j] = math.trunc(img[i,j] * 1.7) # 1 -> 1.7 (실수를 정수화 하기)
        else:
            ex2_result[i,j] = math.trunc(img[i,j] * 1.7) - 255 - 1 # 256 = 0 / 257 = 1
```

```python
plt.imshow(ex2_result, cmap = 'gray')
```

```python
# numpy를 사용하여 랩핑 구현하기(다음 페이지의 '행렬 및 성질' 참고)
# add(100), subtract(100), multiply(1.7), divide(3)

ex2_result1 = (img + 100).astype('uint8') # cv2.add(img, 100)
ex2_result2 = (img - 100).astype('uint8')
ex2_result3 = (img * 1.7).astype('uint8')
ex2_result4 = (img / 3).astype('uint8')

plt.subplot(2, 3, 1); plt.imshow(img, cmap = 'gray')
plt.subplot(2, 3, 2); plt.imshow(ex2_result1, cmap = 'gray')
plt.subplot(2, 3, 3); plt.imshow(ex2_result2, cmap = 'gray')
plt.subplot(2, 3, 4); plt.imshow(ex2_result3, cmap = 'gray')
plt.subplot(2, 3, 5); plt.imshow(ex2_result4, cmap = 'gray')
```
